### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# I matplotlib finns det figur, axlar, axel och artist. 
# Figure är som en duk där dina axlar och axel visas dvs en bas för helheten av koden som du kör.
# Axlar är innehåller y-axeln och x-axeln  som går igenom data punkter samt rubriker. Kort så går axlar igenom plottar eller ploten.
# Axeln är Y-axeln och X-axeln som går i en diagram och visar dimesionerna i diagrammet. 
# En artist är sammanfatningsvis den synliga delen av linjer och former samt texter. Allt som är i en plot är artist. 
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 10)
y = x ** 2
plt.plot(x, y)
plt.title("Pyplot")
plt.xlabel("x-axeln")
plt.ylabel("y-axeln")
plt.show()

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Objekt Orienterad")
ax.set_xlabel("x-axeln")
ax.set_ylabel("y-axeln")
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.scatter(random_data_x, random_data_y, color="blue", alpha=0.5)
ax1.plot(x, y, color="red")
ax1.set_title("Scatter")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.bar(names, values, color="orange")
ax2.set_title("Bar")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Kvantitet")
plt.tight_layout()
plt.show()
```
